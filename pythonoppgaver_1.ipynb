
Feltet i et punkt R fra en punktladning er  $|E| = \frac{Q}{4\pi\varepsilon_0 R^2}$ og går i rettning vekk fra punktladningen

Vi kan skrive det som 

$ E_x = \frac{xQ}{4\pi\varepsilon_0 R^3}$

$ E_y = \frac{yQ}{4\pi\varepsilon_0 R^3}$

Hvor R er avstaden fra ladningen til punktet vi måler, x er avstanden langs x-aksen, og y er avstanden langs y aksen 

For en ladning plassert i \[-1.23, 0.54\] kan vi vise feltet med piler på følgende måte:



````python
import numpy as np
import matplotlib.pyplot as plt
elektrisk_konst = 8.8541878128*10**-12 # €0 med enhet F/m

#grenser for plot
x_lim = 5
y_lim = 5
n = 20 # antallet datapunkter langs X_grid, Y_grid

# gitter:
x_vektor = np.linspace(-x_lim, x_lim, n)
y_vektor = np.linspace(-y_lim, y_lim, n)
X_grid = x_vektor[np.newaxis, :] * np.ones((n,n))
Y_grid = y_vektor[:, np.newaxis] * np.ones((n,n))

# elektrisk felt9
Q_ladning = 1 # ladning til punktladningen, i coulomb 
x_ladning = -1.23 # x-koordinatet til ladningen, i meter
y_ladning = 0.54 # y-koordinatet til ladningen, i meter
R_ladning = np.sqrt( (X_grid-x_ladning)**2 + (Y_grid-y_ladning)**2 ) # avstanden fra ladningen til gitterpunktene beskrevet av X og Y
Ex_ladning = (X_grid-x_ladning) * Q_ladning / (4 * np.pi * elektrisk_konst * R_ladning**3)
Ey_ladning = (Y_grid-y_ladning) * Q_ladning / (4 * np.pi * elektrisk_konst * R_ladning**3)


fig, ax = plt.subplots()
ax.quiver(X_grid, Y_grid, Ex_ladning, Ey_ladning)
ax.set_aspect('equal', adjustable='box') # vi må sørge for at 1 meter langs x-aksen er like langt som 1 meter langs y-aksen.

fig.show()
````

I kompendiet til Skaar står det at kapasitansen (per lengdeenhet) til en koaksialkabel er $ C' = \frac{2\pi\varepsilon}{\log(b/a)} $ (ligning 2.139)

Energien $E$ til en kondensator er lik $\frac{Q'^2}{C'}$ hvor $Q'$ er ladningen, og $C'$ er capasitansen. (alt  per lengdeenhet)

med $a = 2$, $b= 4$, og $Q' = 1$ får vi:

```
C = 2*np.pi*elektrisk_konst/np.log(4/2) # = 8.026073586197363e-11
E = 1**2/C/2 # = 6229696184.9
```
